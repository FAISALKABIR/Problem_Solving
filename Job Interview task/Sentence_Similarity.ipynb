```python
# Load the model
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("bert-base-nli-mean-tokens")
```

```python
# Make an array of all the Questions
df_faq = df["Question"].unique()
df_faq
```

```python
# Make an array of all the Answers
df_faq_ans = df["Answer"].unique()
df_faq_ans
```

```python
# Make an array of all the Test Questions
df_faq_test = test_df["Question"].unique()
df_faq_test
```

```python
# Using sentences encode
df_faq_vec = model.encode(df_faq)
df_faq_vec
```

```python
df_faq_test_vec = model.encode(df_faq_test)
df_faq_test_vec
```

```python
# Check the vector shape
print(df_faq_vec.shape)
print(df_faq_test_vec.shape)
```

```python
# Check the cosine similarity of two sentence
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity([df_faq_test_vec[2]], [df_faq_vec[5]])
```

```python
for i in range(len(df_faq_test)):
    print("Test Question: ", df_faq_test[i])
    cos_list = []
    for j in range(len(df_faq)):
        cos_sim = cosine_similarity([df_faq_test_vec[i]], [df_faq_vec[j]])
        cos_list.append(cos_sim[0][0])

    max_sim = max(cos_list)
    index = cos_list.index(max_sim)
    print(max_sim)
    print(df.loc[[index]])
    print()
```
